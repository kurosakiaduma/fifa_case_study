```python
def open_fifa_version_df(fifa_version):
    
    df = pd.read_csv('./players_' + str(fifa_version) +'.csv', low_memory=False)

    for value in range(15,23):
        if value == fifa_version:
            df['fifa_version'] = int(fifa_version)
        
    new_cols_order = ['fifa_version'] + df.columns.to_list()[:-6]
    # remove the picture URLs
    
    df = df[new_cols_order]
    #21defending_awareness, 21attacking etc
    
    return df
```

```python
players_df = pd.DataFrame()

for fifa_version in range(15,23):
    
    #players_df = players_df.append(open_fifa_version_df(fifa_version)) -- this method has been deprecated since
    
    players_df = pd.concat([players_df, open_fifa_version_df(fifa_version)])
    print(len(players_df))
    
    players_df
```
